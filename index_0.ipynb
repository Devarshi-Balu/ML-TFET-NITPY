```python
import pandas as pd 

df = pd.read_excel("dataset.xlsx")

from sklearn.model_selection import train_test_split



import sys 
sys.path.append("C:/Users/studi/Desktop/machine learning/_ ml _ internship/global")
```

```python
df.describe()
print(df.columns)

features = ['Tsi', 'Tox', 'Ns', 'Nc', 'Nd', 'Vds', 'Vgs']
print(features)
target = 'Id'
print(target)
```

```python

X = df[features]
y = df[target]


X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.000000000000000001,      # 20% of data for testing
                                                    random_state=42,    # For reproducibility
                                                    shuffle=True) 



from functions import x_test, y_test 
X_test = x_test
y_test = y_test
```

```python

from random import Random
from sklearn.preprocessing import StandardScaler, RobustScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both the training and test data

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)



from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators = 100, criterion="squared_error", random_state=42, max_depth = None, n_jobs=-1)
```

```python
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))

print(r2_score(y_test[0:203], y_pred[0:203]))


```

```python
import matplotlib.pyplot as plt 

Vgs = X_test['Vgs']

plt.semilogy(Vgs.argsort(), y_test.argsort(), color = "red")
plt.semilogy(Vgs.argsort(), y_pred.argsort(), color="blue")
plt.show()

```
