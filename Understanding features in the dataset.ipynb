```python
import pandas as pd
df_train = pd.read_csv('main_dataset.csv')

```

```python
print(df_train.columns)
```

```python
features = ['Tsi', 'Tox', 'Ns', 'Nc', 'Nd', 'Vds', 'Vgs']
target = 'Id'\

# df_train = df_train.iloc[0:203]
```

```python
import matplotlib.pyplot as plt 

for x in features :
    x_value = df_train[x]
    y_value = df_train[target]

    # plt.plot(x_value)
    plt.scatter(x_value, y_value, color = 'green')
    plt.title(f'{x} vs {target}')
    # plt.legend()
    plt.show()

    
```
